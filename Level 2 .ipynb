```python
#  Plot a histogram to visualize the distribution of ratings
plt.figure(figsize=(10, 6))
plt.hist(data['Aggregate rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.grid(axis='y')
```

```python
#  Identify the most common rating range
rating_bins, bin_edges = pd.cut(data['Aggregate rating'], bins=10, retbins=True)  
rating_counts = rating_bins.value_counts().sort_index()  
most_common_rating_range = rating_counts.idxmax()  
count_in_most_common_range = rating_counts.max()  

```

```python
print(f'The most common rating range is {most_common_rating_range} with {count_in_most_common_range} restaurants.')
```

```python
data = pd.DataFrame({
    'Restaurant Name': ['McDonald\'s', 'Subway', 'Pizza Hut', 'McDonald\'s', 'Burger King', 'Subway', 'Starbucks', 'Pizza Hut', 'McDonald\'s'],
    'Aggregate Rating': [4.2, 3.8, 4.0, 4.1, 3.9, 4.0, 4.5, 3.7, 4.3],
    'Rating text': ['Good', 'Good', 'Very Good', 'Good', 'Excellent', 'Very Good', 'Good', 'Excellent', 'Very Good']
})
```

```python
# Convert restaurant names to lowercase for case-insensitive matching
data['Restaurant Name'] = data['Restaurant Name'].str.lower()

# Identify restaurant chains based on frequency of occurrence
restaurant_counts = data['Restaurant Name'].value_counts()

# Filter chains (restaurants with multiple locations)
chains = restaurant_counts[restaurant_counts > 1]

# Print out the identified chains
if len(chains) > 0:
    print("Identified restaurant chains:")
    for chain, count in chains.items():
        print(f"{chain.capitalize()} - {count} locations")
else:
    print("No restaurant chains identified in the dataset.")
    
```

#  Determine if certain cuisine combinations tend to have higher ratings.


```python
import pandas as pd
from collections import Counter
```

```python
data =pd.read_csv("Data_set.csv")
```

```python
data
```

```python

# Assuming the column with cuisines is named 'Cuisines' and ratings are in 'Rating'
# If cuisines are listed as comma-separated values:
def normalize_cuisines(cuisine_list):
    if isinstance(cuisine_list, str):
        return [cuisine.strip() for cuisine in cuisine_list.split(',')]  # Convert string to a list
    else:
        return [cuisine.strip() for cuisine in cuisine_list]  # Handle non-strings

# Create or load your DataFrame (replace with your data loading code)
data = pd.DataFrame({'Cuisines': ['Italian,Asian', 'Mexican', 'American,French', 'Thai,Chinese'],
                     'Aggregate rating': [4.5, [3.8, 3.9], 5.0, 4.2]})

# Normalize cuisine lists
data['Cuisines'] = data['Cuisines'].apply(normalize_cuisines)

# Sort cuisines within each list for consistent counting (optional)
data['Cuisines'] = data['Cuisines'].apply(lambda x: tuple(sorted(x)))

# Handle single ratings (if applicable):
def handle_rating(rating):
    if isinstance(rating, list):
        return rating[0]  # Use the first rating in the list
    else:
        return rating  # Keep single ratings as is



data['Aggregate rating'] = data['Aggregate rating'].apply(handle_rating)


# Combine cuisine and rating into tuples for counting
cuisine_rating_pairs = zip(data['Cuisines'], data['Aggregate rating'])

# Count the frequency of each unique combination of cuisines and ratings
cuisine_combinations = Counter(cuisine_rating_pairs)

# Calculate average rating for each cuisine combination
avg_ratings = {combo: ratings for combo, ratings in cuisine_combinations.items()}  # Keep all entries


# Find combinations with high average rating (adjust threshold as needed)
high_rated_combos = {combo: rating for combo, rating in avg_ratings.items() if rating >= 4.3}

# Print the results
print("Cuisine Combinations with High Average Rating (>= 4.3):")
for combo, rating in high_rated_combos.items():
    cuisines, _ = zip(*combo)  # Unzip cuisine and rating
    print(f"{', '.join(cuisines)}: Average Rating = {rating:.1f}")
```
