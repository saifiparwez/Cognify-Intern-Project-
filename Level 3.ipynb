# Analyze if there is a correlation between the number of votes and the rating of a restaurant.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
data=pd.read_csv("Data_set.csv")
```

```python
data
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
# Assuming the columns are named 'votes' and 'rating'
# Replace these with the actual column names if they are different

# Drop rows with missing values in 'votes' or 'rating'
df = data.dropna(subset=['Votes', 'Aggregate rating'])

# Calculate the Pearson correlation coefficient
correlation, p_value = pearsonr(data['Votes'], data['Aggregate rating'])

print(f'Pearson correlation coefficient: {correlation}')
print(f'P-value: {p_value}')

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df)
plt.title('Scatter Plot of Votes vs. Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate rating')
plt.show()
```

# Task 3: Price Range vs. Online Delivery and Table Booking


# Analyze if there is a relationship between the price range and the availability of online delivery and table booking.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
```

```python
data = pd.read_csv("Data_set.csv")
```

```python
data
```

```python
# Assuming the columns are named 'price_range', 'online_delivery', and 'table_booking'
# Replace these with the actual column names if they are different

# Ensure the availability columns are categorical
data['Has Online delivery'] = data['Has Online delivery'].astype('category')
data['Has Table booking'] = data['Has Table booking'].astype('category')

# Cross-tabulation for price range and online delivery
cross_tab_delivery = pd.crosstab(data['Price range'], data['Has Online delivery'])
print("\nCross-tabulation for Price Range and Online delivery:")
print(cross_tab_delivery)

# Chi-square test for independence between price range and online delivery
chi2, p, dof, ex = chi2_contingency(cross_tab_delivery)
print(f"\nChi-square test for Price range and Online delivery: chi2 = {chi2}, p-value = {p}")

# Cross-tabulation for price range and table booking
cross_tab_booking = pd.crosstab(data['Price range'], data['Has Table booking'])
print("\nCross-tabulation for Price Range and Table booking:")
print(cross_tab_booking)

# Chi-square test for independence between price range and table booking
chi2, p, dof, ex = chi2_contingency(cross_tab_booking)
print(f"\nChi-square test for price range and table booking: chi2 = {chi2}, p-value = {p}")
```

```python
# Visualize the relationship with heatmaps
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(cross_tab_delivery, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Price Range vs Online Delivery')
plt.xlabel('Online Delivery')
plt.ylabel('Price Range')

plt.subplot(1, 2, 2)
sns.heatmap(cross_tab_booking, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Price Range vs Table Booking')
plt.xlabel('Table Booking')
plt.ylabel('Price Range')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm  # Importing statsmodels.api for logistic regression
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Data_set.csv')

# Assuming the columns are named 'Price range', 'Has Online delivery', and 'Has Table booking'
# Replace these with the actual column names if they are different

# Convert categorical columns to numeric values
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})
data['Has Table booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})
data['Price range'] = data['Price range'].astype('category').cat.codes

# Drop rows with missing values in any of the relevant columns
data = data.dropna(subset=['Has Online delivery', 'Has Table booking', 'Price range'])

# Ensure data is numeric
data['Has Online delivery'] = pd.to_numeric(data['Has Online delivery'])
data['Has Table booking'] = pd.to_numeric(data['Has Table booking'])
data['Price range'] = pd.to_numeric(data['Price range'])

# Logistic regression for online delivery
X_delivery = data[['Price range']]
X_delivery = sm.add_constant(X_delivery)  # Adding a constant term for the intercept
y_delivery = data['Has Online delivery']
model_delivery = sm.Logit(y_delivery, X_delivery)
result_delivery = model_delivery.fit()

print("\nLogistic Regression Results for Online Delivery:")
print(result_delivery.summary())

# Logistic regression for table booking
X_booking = data[['Price range']]
X_booking = sm.add_constant(X_booking)  # Adding a constant term for the intercept
y_booking = data['Has Table booking']
model_booking = sm.Logit(y_booking, X_booking)
result_booking = model_booking.fit()

print("\nLogistic Regression Results for Table Booking:")
print(result_booking.summary())

# Visualization
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='Price range', y='Has Online delivery', data=data, ci=None)
plt.title('Price Range vs Online Delivery')
plt.xlabel('Price Range')
plt.ylabel('Proportion Offering Online Delivery')

plt.subplot(1, 2, 2)
sns.barplot(x='Price range', y='Has Table booking', data=data, ci=None)
plt.title('Price Range vs Table Booking')
plt.xlabel('Price Range')
plt.ylabel('Proportion Offering Table Booking')

plt.tight_layout()
plt.show()
```
